# Tuplas

* Uma tupla é uma sequência imutável e de tamanho fixo de objetos Python que, uma vez atribuída, não pode ser alterada. A maneira mais fácil de criar uma tupla é com uma sequência de valores separados por vírgula inseridos em parênteses:

```python
tup = (4, 5, 6)
tup
```

* Podemos converter qualquer sequência ou iterador em uma tupla chamando tuple():

```python
tuple([3,9,4])
```

```python
tup_string = tuple('string')
tup_string
```

* Tuplas aninhadas:

```python
nested_tup = (4, 5, 6), (7, 8)
nested_tup
```

* Acesso por indexação encadeada:

```python
nested_tup[0]
```

```python
nested_tup[1]
```

```python
nested_tup[0][0]
```

```python
nested_tup[1][1]
```

* Embora os objetos armazenados em uma tupla possam ser mutáveis, após a tupla ser criada não será possível modificar que objeto ficará armazenado em cada posição:

```python
tup = tuple(['foo', [1,2], True])
# tup[2] = False 
tup[1].append(3)
tup
```
