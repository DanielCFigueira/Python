# Listas

* Ao contrário das tuplas, as listas têm tamanho variável e seu conteúdo pode ser modificado diretamente no local. As listas são mutáveis. Você pode defini-las usando colchetes ([ ]) ou empregando a função de tipo list:

```python
# Criando uma lista:
a_lista = [2, 3, 7, None]

# Você também pode usar a função list() para criar ou converter outra estrutura de dados para uma lista:
tup = ("Daniel", "Max", "Magnus")
b_lista = list(tup)

print(a_lista, b_lista)
```

* Você também pode alterar um elemento que está na lista. Cada objeto na lista recebe um indíce, logo, basta você colocar entre colchetes o indíce do objeto que você quer mudar ou excluir para que você possa efetuar a operação:

```python
a_lista[3] = 12
a_lista
```

* As listas e tuplas são semanticamente semelhantes (embora as tuplas não possam ser modificadas) e podem ser usadas de maneira intercambiável em muitas funções.
Geralmente, a função interna list é usada no processamento de dados como
uma maneira de materializar um iterador ou uma expressão geradora:

```python
gen = range(10)
print(gen)

list(gen)
```

# Inclusão e Remoção de Elementos

* Elementos podem ser acrescentados ao fim da lista com o método append:

```python
b_lista.append("Figueira")
b_lista
```

* Usando insert você pode inserir um elemento em um local específico da lista:

```python
a_lista.insert(3, 8)
a_lista
```

* A operação inversa a insert é pop, que remove e retorna um elemento de um índice específico:

```python
b_lista.pop(3)
b_lista
```

* Elementos podem ser removidos pelo valor com remove, que localiza a primeira ocorrência do valor e a remove da lista:

```python
a_lista.append(2)
a_lista
```

```python
a_lista.remove(2)
a_lista
```

* Para verificar se uma lista contém um valor, use a palavra-chave in:

```python
"Daniel" in b_lista
```

* A palavra-chave not pode ser usada como negação de in: 

```python
"Daniel" not in b_lista
```

* Verificar se uma lista contém um valor é muito mais lento do que fazer o mesmo com dicionários e conjuntos, já que o Python faz uma verificação linear nos valores da lista, enquanto nas outras estruturas (baseadas em tabelas hash) a verificação ocorre em tempo constante.

# Concatenação e Combinação de Listas

* Como nas tuplas, somar duas listas com + as concatena:

```python
lista1 = [4, None, "foo"]
lista2 = [7, 8, (2,3)]
sum_lista = lista1 + lista2
sum_lista
```

* Se você tiver uma lista já definida, poderá acrescentar vários elementos a ela usando o método extend:

```python
x = [4, 5, 0, "Daniel"]
x.extend(["Max", "Magnus", 14, 44, 27, (9, 0)])
x
```

* É bom ressaltar que a concatenação de listas por soma é uma operação comparativamente dispendiosa já que uma nova lista deve ser criada e os objetos devem ser copiados nela. Geralmente é preferível usar extend para o acréscimo de elementos a uma lista existente, principalmente se você estiver criando uma lista grande. Portanto:

```python
lista = ["Daniel", "Max", "Magnus"]
tudo = []
for i in lista:
    tudo.extend(i)
tudo
```

* É mais rápido do que a alternativa pela concatenação:

```python
lista = ["Boa", "Noite", "Bixo"]
tudo = [] 
for i in lista:
    tudo = tudo + lista
tudo
```

# Ordenação

* Você pode ordenar uma lista diretamente no local(sem criar um novo objeto) chamando sua função sort: 

```python
a = [7, 2, 5, 1, 3]
a.sort()
a
```

* O sort tem algumas opções que ocasionalmente são úteis. Uma é a possibilidade de passar uma chave de ordenação secundária – isto é, uma função que produza um valor a ser usado para ordenar os objetos. Por exemplo, poderíamos ordenar uma coleção de strings pelos seus tamanhos:

```python
b = ["saw", "small", "He", "foxes", "six"]
b.sort(key=len)
b
```

# Fatiamento

* Você pode selecionar seções da maioria dos tipos de sequência usando a notação de fatias (slices), que em sua forma básica é composta da passagem de start:stop (início:fim) para o operador de indexação []:

```python
seq = [7, 2, 3, 7, 5, 6, 0, 1]
seq[1:5]
```

* As fatias também podem receber uma sequência como atribuição: 

```python
seq[3:5] = [6, 3]
seq
```

* Embora o elemento do índice start seja incluído, o índice stop não é incluído, de modo que o número de elementos do resultado é stop - start.

* Podemos omitir start ou stop, caso em que por padrão eles serão o início e o fim da sequência, respectivamente: 

```python
seq[:5]
```

```python
seq[3:]
```

* Índices negativos fatiam a sequência em relação ao final:

```python
seq
```

```python
seq[-4:]
```

```python
seq[-6:-2]
```

* Um step (intervalo) também poderia ser usado após um segundo símbolo de dois pontos para, por exemplo, obtermos elementos alternados:

```python
seq[::2]
```

* Um uso inteligente desse recurso seria a passagem de -1, o que teria o útil efeito de inverter uma lista ou tupla:

```python
seq
```

```python
seq[::-1]
```
